```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = True
```

```python
def fac(n):
    if n<0:
        raise ValueError
    elif n==0:
        return 1
    else:
        return n*fac(n-1)
```

```python
def arcsin(x, n):
    return sum([fac(2*k)/(4**k * fac(k)**2*(2*k+1)) * x**(2*k+1) for k in range(0, n+1)])
```

```python
def arctan(x, n):
    return sum([(-1)**k/(2*k+1) * x**(2*k+1) for k in range(0, n+1)])
```

```python
N = 20
Ns = np.arange(0, N+1)
pi_asin = 6 * np.array([arcsin(0.5, n) for n in range(0, N+1)])
pi_atan = 4 * np.array([arctan(1.0, n) for n in range(0, N+1)])
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 5))
ax.set_title(r"$\pi$ from two Maclaurin series", size=30)
ax.set_xlabel("$n$", size=20)
ax.set_ylabel("$T_{n}$", rotation=0, size=20)
ax.tick_params(labelsize=15)
ax.plot(Ns, pi_asin, "o", label=r"$6\arcsin\left(\frac{1}{2}\right)$")
ax.plot(Ns, pi_atan, "o", label=r"$4\arcsin\left(1\right)$")
ax.plot(Ns, np.ones(N+1)*np.pi, label=r"$\pi$")
ax.legend()
```
